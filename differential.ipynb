## Example $\mathbb{R} \to \mathbb{R}$
--- 

$$ df = f'(x)dx $$

##### Numerical example

$$ f(x) = x^3 $$
$$ f'(x) = 3x^2 $$
$$ x = 2 $$
$$ dx = 0.01 $$
$$ f'(2) = 12 $$
$$ df  = 0.12 $$

##### Verification via finite difference

$$ f(x) = 8 $$
$$ f(x + dx) \approx 8.12 $$
$$ f(x + dx) - f(x) = 0.12 \approx df $$

## Example $\mathbb{R}^{n} \to \mathbb{R}$
---
$$ df = \Big( \nabla f(\mathbf{x}) \Big)^\top \cdot d\mathbf{x} $$

$$ f(\mathbf{x}) = f(x_1, x_2, \ldots, x_n) $$

$$ \nabla f(\mathbf{x}) = 
\begin{bmatrix} 
    \frac{\partial f}{\partial x_1} \\
    \frac{\partial f}{\partial x_2} \\
    \vdots \\
    \frac{\partial f}{\partial x_n} 
\end{bmatrix} $$

##### Numerical example

$$ f(\mathbf{x}) = \sum_{i} x_{i}^2 $$

$$ \nabla f(\mathbf{x}) = 
\begin{bmatrix} 
    2x_1 \\
    2x_2 \\
    \vdots \\
    2x_n
\end{bmatrix} = 
2\mathbf{x} $$

$$
\mathbf{x} =
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix}
$$

$$
d\mathbf{x} = \begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix}
$$

$$ df = [2, 4, 6] \cdot \begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix} = 0.12 $$

##### Verification via finite difference

$$ f(\mathbf{x}) =14 $$

$$ f(\mathbf{x} + d\mathbf{x}) = 14.12 $$

$$ f(\mathbf{x} + d\mathbf{x}) - f(\mathbf{x}) = 0.12 \approx df $$

## Example $\mathbb{R} \to \mathbb{R}^{n}$
---
$$ dF = J_F(x) \cdot dx $$

$$
F(x) = 
\begin{bmatrix}
f_1(x) \\ f_2(x) \\ f_3(x)
\end{bmatrix}
$$

$$ f_1, f_2, \cdots, f_n: \mathbb{R} \to \mathbb{R} $$

$$ J_F (x) = 
\begin{bmatrix} 
\frac{\partial f_1}{\partial x} \\ 
\frac{\partial f_2}{\partial x} \\
\frac{\partial f_3}{\partial x} 
\end{bmatrix}
$$

##### Numerical example

$$
F(x) =
\begin{bmatrix}
x \\ x^2 \\ x^3
\end{bmatrix}
$$

$$ J_F (x) = 
\begin{bmatrix} 
1 \\ 
2x \\
3x^2
\end{bmatrix}
$$

$$ x = 2 $$

$$ dx = 0.01 $$

$$ 
dF = 
\begin{bmatrix} 1 \\ 2 \cdot 2 \\ 3 \cdot 2^2 \end{bmatrix} \cdot 0.01 =
\begin{bmatrix} 0.01 \\ 0.04 \\ 0.12 \end{bmatrix} 
$$

##### Verification via finite difference

$$ f(x) = \begin{bmatrix} 2 \\ 4 \\ 8 \end{bmatrix} $$

$$ f(x + dx) =\begin{bmatrix} 2.01 \\  4.04 \\ 8.12 \end{bmatrix} $$

$$ 
f(x + dx) - f(x) = 
\begin{bmatrix} 0.01 \\  0.04 \\ 0.12 \end{bmatrix} \approx 
df
$$

## Example $\mathbb{R}^{n} \to \mathbb{R}^{n}$
---
$$ dF = J_F(\mathbf{x}) \cdot d\mathbf{x} $$

$$ 
F(\mathbf{x}) = 
    \begin{bmatrix} 
        f_1(\mathbf{x}) \\ 
        f_2(\mathbf{x}) \\
        \vdots \\ 
        f_n(\mathbf{x}) 
    \end{bmatrix} 
$$

$$ f_1, f_2, \cdots, f_n: \mathbb{R}^n \to \mathbb{R} $$

$$ J_F(\mathbf{x}) = 
\begin{bmatrix} 
    \Big( \nabla f_1(\mathbf{x}) \Big)^T \\ 
    \Big( \nabla f_2(\mathbf{x}) \Big)^T \\ 
    \vdots \\ 
    \Big( \nabla f_n(\mathbf{x}) \Big)^T 
\end{bmatrix} =
\begin{bmatrix} 
    \frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\
    \frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\
    \vdots & \vdots & \ddots & \vdots \\
    \frac{\partial f_n}{\partial x_1} & \frac{\partial f_n}{\partial x_2} & \cdots & \frac{\partial f_n}{\partial x_n}
\end{bmatrix}
$$

##### Numerical example

$$ 
F(\mathbf{x}) =
\begin{bmatrix} 
    x_1 \cdot \sum(\mathbf{x}) \\ 
    x_2 \cdot \sum(\mathbf{x}) \\ 
    \vdots \\ 
    x_n \cdot \sum(\mathbf{x}) 
\end{bmatrix}
$$

$$ f_i(\mathbf{x}) = x_i \sum\mathbf{x}$$

$$
\frac{\partial f_1}{\partial x_1} = \sum(\mathbf{x}) + x_1, \,
\frac{\partial f_1}{\partial x_2} = x_1, \cdots
$$

$$
\frac{\partial f_i}{\partial x_j} = \delta_{ij} \sum_{k}{x_k} + x_i
$$

$$
J_F(\mathbf{x}) =
\begin{bmatrix} 
    \sum(\mathbf{x}) + x_1 & x_1 & \cdots & x_1 \\ 
    x_2 & \sum(\mathbf{x}) + x_2 & \cdots & x_2 \\ 
    \vdots & \vdots & \ddots & \vdots \\ 
    x_n & x_n & \cdots & \sum(\mathbf{x}) + x_n 
\end{bmatrix}
$$

$$ \mathbf{x} = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} $$

$$ d\mathbf{x} = \begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix} $$

$$ 
dF = 
\begin{bmatrix} 
    6 + 1 & 1 & 1 \\ 
    2 & 6 + 2 & 2 \\ 
    3 & 3 & 6 + 3 
\end{bmatrix} \cdot
\begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix} =
\begin{bmatrix} 0.08 \\ -0.08 \\ 0.24 \end{bmatrix} 
$$

##### Verification via finite difference

$$ F(\mathbf{x}) = \begin{bmatrix} 6 \\ 12 \\ 18 \end{bmatrix} $$

$$ F(\mathbf{x} + d\mathbf{x}) = \begin{bmatrix} 6.08 \\ 11.91 \\ 18.24 \end{bmatrix} $$

$$ 
F(\mathbf{x} + d\mathbf{x}) - F(\mathbf{x}) = 
\begin{bmatrix} 0.08 \\ -0.08 \\ 0.24 \end{bmatrix} \approx dF 
$$

## Example $\mathbb{R}^{n} \to \mathbb{R}^{n} - \text{element-wise}$

---

$$ 
dF = J_F(\mathbf{x}) \cdot d\mathbf{x} \overset{\text{element-wise}}{=} 
\text{diag}(J_F(\mathbf{x})) \odot d\mathbf{x} 
$$

$$ 
F(\mathbf{x}) = 
\begin{bmatrix} f_1(\mathbf{x}) \\ f_2(\mathbf{x}) \\ \vdots \\ f_n(\mathbf{x}) \end{bmatrix} 
$$

$$ f_1, f_2, \cdots, f_n: \mathbb{R}^n \to \mathbb{R} $$

$$ J_F(\mathbf{x}) = 
\begin{bmatrix} 
    \Big( \nabla f_1(\mathbf{x}) \Big)^{\top} \\ 
    \Big( \nabla f_2(\mathbf{x}) \Big)^{\top} \\ 
    \vdots \\ 
    \Big( \nabla f_n(\mathbf{x}) \Big)^{\top} 
\end{bmatrix} \overset{\text{element-wise}}{=}
\begin{bmatrix} 
    \frac{\partial f_1}{\partial x_1} & 0 & \cdots & 0 \\
    0 & \frac{\partial f_2}{\partial x_2} & \cdots & 0 \\
    \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & \cdots & \frac{\partial f_n}{\partial x_n}
\end{bmatrix}
$$

##### Numerical example

$$ F(\mathbf{x}) = 2\mathbf{x}^3 $$

$$ f_i(\mathbf{x}) = 2x_i^3 $$

$$ 
\frac{\partial f_i}{\partial x_j} = 
\begin{cases} 6x_i^2 & \text{if } i = 
j \\ 0 & \text{if } i \neq j \end{cases} 
$$

$$ 
J_F(\mathbf{x}) = 
\begin{bmatrix} 
6x_1^2 & 0 & \cdots & 0 \\ 
0 & 6x_2^2 & \cdots & 0 \\ 
\vdots & \vdots & \ddots & \vdots \\ 
0 & 0 & \cdots & 6x_n^2 
\end{bmatrix} 
$$

$$ \mathbf{x} = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} $$

$$ d\mathbf{x} = \begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix} $$

$$
dF = 
\begin{bmatrix} 
    6 \cdot 1^2 & 0 & 0 \\ 
    0 & 6 \cdot 2^2 & 0 \\ 
    0 & 0 & 6 \cdot 3^2 
\end{bmatrix} \cdot 
\begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix}
=
\begin{bmatrix} 
    6 \\ 
    24 \\ 
    54 
\end{bmatrix} \odot 
\begin{bmatrix} 0.01 \\ -0.02 \\ 0.03 \end{bmatrix} =
\begin{bmatrix} 0.06 \\ -0.48 \\ 1.62 \end{bmatrix} 
$$

##### Verification via finite difference

$$ F(\mathbf{x}) = \begin{bmatrix} 2 \\ 16 \\ 54 \end{bmatrix} $$

$$ F(\mathbf{x} + d\mathbf{x}) = \begin{bmatrix} 2.06 \\ 15.52 \\ 55.62 \end{bmatrix} $$

$$ F(\mathbf{x} + d\mathbf{x}) - F(\mathbf{x}) = \begin{bmatrix} 0.06 \\ -0.48 \\ 1.62 \end{bmatrix} \approx dF $$

## Example $\mathbb{R} \times \mathbb{R} \to \mathbb{R}$

---

$$ df = \frac{\partial f}{\partial x_1}(x_1, x_2)\, dx_1 \;+\; \frac{\partial f}{\partial x_2}(x_1, x_2)\, dx_2 $$

$$ f(x_1, x_2) = x_1^2 + x_2^3 $$
$$ \frac{\partial f}{\partial x_1} = 2x_1 $$
$$ \frac{\partial f}{\partial x_2} = 3x_2^2 $$

##### Numerical example

$$ x_1 = 2 $$
$$ x_2 = 3 $$

$$ dx_1 = 0.01 $$
$$ dx_2 = -0.02 $$

$$ \frac{\partial f}{\partial x_1}(2,3) = 4 $$
$$ \frac{\partial f}{\partial x_2}(2,3) = 27 $$

$$ df = 4 \cdot 0.01 \; + \; 27 \cdot (-0.02) = -0.50 $$

##### Verification via finite difference

$$ f(\mathbf{x}, \mathbf{y}) = 2^2 + 3^3 = 31 $$
$$ f(\mathbf{x} + d\mathbf{x},\; \mathbf{y} + d\mathbf{y}) \approx 30.5040 $$

$$ f(\mathbf{x} + d\mathbf{x}, \mathbf{y} + d\mathbf{y}) - f(\mathbf{x}, \mathbf{y})
\approx 30.5040 - 31
\approx -0.496 \approx df $$

## Example $\mathbb{R}^n \times \mathbb{R}^n \to \mathbb{R}$

--- 

$$ df
= \big( \nabla_{\mathbf{x}} f(\mathbf{x}, \mathbf{y}) \big)^\top d\mathbf{x} \; + \;
\big( \nabla_{\mathbf{y}} f(\mathbf{x}, \mathbf{y}) \big)^\top d\mathbf{y} 
$$

$$ \nabla_{\mathbf{x}} f(\mathbf{x}, \mathbf{y}) =
\begin{bmatrix}
    \frac{\partial f}{\partial x_1} \\
    \frac{\partial f}{\partial x_2} \\
    \vdots \\
    \frac{\partial f}{\partial x_n} 
\end{bmatrix} 
$$

$$ \nabla_{\mathbf{y}} f(\mathbf{x}, \mathbf{y}) =
\begin{bmatrix}
    \frac{\partial f}{\partial y_1} \\
    \frac{\partial f}{\partial y_2} \\
    \vdots \\
    \frac{\partial f}{\partial y_n} 
\end{bmatrix} 
$$

##### Numerical example

$$ f(\mathbf{x}, \mathbf{y}) = \sum_{i} x_i y_i $$

$$ 
\nabla_{\mathbf{x}} f(\mathbf{x}, \mathbf{y}) =
\begin{bmatrix}
    y_1 \\
    y_2 \\
    \vdots \\
    y_n
\end{bmatrix} = \mathbf{y} 
$$

$$ 
\nabla_{\mathbf{y}} f(\mathbf{x}, \mathbf{y}) =
\begin{bmatrix}
    x_1 \\
    x_2 \\
    \vdots \\
    x_n
\end{bmatrix} = \mathbf{x} 
$$

$$ 
\mathbf{x} =
\begin{bmatrix}
    1 \\
    2 \\
    3
\end{bmatrix} 
$$

$$ 
\mathbf{y} =
\begin{bmatrix}
    4 \\
    5 \\
    6
\end{bmatrix} 
$$

$$ 
d\mathbf{x} =
\begin{bmatrix}
    0.01 \\
    -0.02 \\
    0.03
\end{bmatrix} 
$$

$$ 
d\mathbf{y} =
\begin{bmatrix}
    0.02 \\
    0.01 \\
    -0.01
\end{bmatrix} 
$$

$$ 
df = [4, 5, 6]
\begin{bmatrix}
    0.01 \\
    -0.02 \\
    0.03
\end{bmatrix} \; + \;
[1, 2, 3]
\begin{bmatrix}
    0.02 \\
    0.01 \\
    -0.01
\end{bmatrix} = 0.13 $$

##### Verification via finite difference

$$ f(\mathbf{x}, \mathbf{y}) = 32 $$

$$ f(\mathbf{x} + d\mathbf{x}, \mathbf{y} + d\mathbf{y}) \approx 32.1297 $$

$$ f(\mathbf{x} + d\mathbf{x}, \mathbf{y} + d\mathbf{y}) - f(\mathbf{x}, \mathbf{y}) \approx 
0.1297 \approx df $$

## Example $\mathbb{R}^{n \times m} \times \mathbb{R}^{n \times m} \to \mathbb{R}$

$$ 
df = \langle \nabla_X f(X, Y),\, dX \rangle_F \;+\;
\langle \nabla_Y f(X, Y),\, dY \rangle_F 
$$

##### Numerical example

$$ f(X, Y) = \mathrm{tr}(XY) $$
$$ \nabla_X f(X, Y) = Y^\top $$
$$ \nabla_Y f(X, Y) = X^\top $$

$$ X =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

$$
Y =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

$$
dX =
\begin{bmatrix}
0.01 & -0.02 \\
0.03 & 0.01
\end{bmatrix}
$$

$$
dY =
\begin{bmatrix}
-0.02 & 0.01 \\
0.00 & -0.03
\end{bmatrix}
$$

$$
\nabla_X f = Y^\top =
\begin{bmatrix}
5 & 7 \\
6 & 8
\end{bmatrix}
$$

$$
\nabla_Y f = X^\top =
\begin{bmatrix}
1 & 3 \\
2 & 4
\end{bmatrix}
$$

$$
df
= \langle Y^\top, dX \rangle_F
\;+\;
\langle X^\top, dY \rangle_F
$$

$$
\langle Y^\top, dX \rangle_F = 0.17
$$

$$
\langle X^\top, dY \rangle_F = -0.11
$$

$$
df = 0.17 - 0.11 = 0.06
$$

##### Verification via finite difference

$$
f(X, Y) = 69
$$

$$
f(X + dX, Y + dY) = 68.96
$$

$$ f(X + dX, Y + dY) - f(X, Y) = 68.96 - 69 = -0.04 $$
