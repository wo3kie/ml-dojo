```python
import torch as tr
import torch.nn as nn
import torch.optim as optim
import torch.autograd as autograd


import import_ipynb
import sigmoid # type: ignore
import linear # type: ignore
import binary_cross_entropy # type: ignore
from common import assert_eq, average_run, test_perceptron_boolean # type: ignore

class Per_Lin_Sig_BCE_Backward(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()

        #
        # For demonstration purposes
        #

        self.linear =  linear.Linear(in_features, out_features)
        self.sigmoid = sigmoid.Sigmoid()

        #
        # [Linear Layer] + [BinaryCrossEntropyWithLogits = Sigmoid + BinaryCrossEntropy] 
        # is more numerically stable than [Linear] + [Sigmoid] + [BinaryCrossEntropy]
        #

    def forward(self, x):
        self.z = self.linear(x)
        self.p = self.sigmoid(self.z)
        return self.p
    
    # Extension
    # To facilitate testing and experimentation various perceptrons.
    def learn(self, x, y):
        return binary_cross_entropy.BinaryCrossEntropy()(self(x), y)
    
    # Extension
    # To facilitate testing and experimentation various perceptrons.
    def predict(self, x):
        with tr.no_grad():
            return sigmoid.sigmoid(self.linear(x)) > 0.5
```

```python

def test_Per_Lin_Sig_BCE_Backward(bool_fn, epochs, lr=0.1):
    return test_perceptron_boolean(Per_Lin_Sig_BCE_Backward(2, 1), bool_fn, epochs=epochs, lr=lr)


if __name__ == "__main__":
    N = 10
    
    logical_and = tr.logical_and
    print(f" AND, {5*N} epochs: {average_run(1*N, lambda: test_Per_Lin_Sig_BCE_Backward(logical_and, epochs=5*N)):.2f}")

    logical_or = tr.logical_or
    print(f"  OR, {5*N} epochs: {average_run(1*N, lambda: test_Per_Lin_Sig_BCE_Backward(logical_or, epochs=5*N)):.2f}")

    logical_nand = lambda x1, x2: tr.logical_not(tr.logical_and(x1, x2))
    print(f"NAND, {5*N} epochs: {average_run(1*N, lambda: test_Per_Lin_Sig_BCE_Backward(logical_nand, epochs=5*N)):.2f}")

    logical_xor = lambda x1, x2: tr.logical_xor(x1, x2)
    print(f" XOR, {5*N} epochs: {average_run(1*N, lambda: test_Per_Lin_Sig_BCE_Backward(logical_xor, epochs=5*N)):.2f}")
```
