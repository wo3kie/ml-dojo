### Example 1

##### Definition of the function

$$ f:\mathbb{R}^{n} \to \mathbb{R} $$

$$ f(x) = ax_1 $$

For demonstration purposes we use a simple scalar field on $\mathbb{R}^n$. The function depends only on the first coordinate, and completely ignores the others.

##### Partial derivatives

$$ \frac{\partial f}{\partial x_i} = \begin{dcases} a & \text{if } i = 1 \\ 0 & \text{otherwise} \end{dcases} $$

##### Gradient  

$$ \nabla f(x) = \begin{bmatrix} \frac{\partial f}{\partial x_1} & \frac{\partial f}{\partial x_2} & \cdots & \frac{\partial f}{\partial x_n} \end{bmatrix} = \begin{bmatrix} a & 0 & \cdots & 0 \end{bmatrix} $$

The gradient always has the same shape as the input vector.
Even if the function uses only one coordinate, the gradient must still be an 𝑛-dimensional vector, because the differential uses an inner product, which requires matching dimensions.

##### Differential  
  
$$ df = \langle \nabla f(x), dx \rangle = a \cdot dx_1 $$

The differential is the linear approximation of the change in the function. Since only $x_1$ matters, only $dx_1$ contributes to the change.

##### Numerical example  
  
$$ a = 5 $$

$$ x = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix} $$

$$ dx = \begin{bmatrix} 0.01 & -0.02 & 0.03 \end{bmatrix} $$

$$ \nabla f(x) = \begin{bmatrix} 5 & 0 & 0 \end{bmatrix} $$

$$ df = \langle \nabla f(x), dx \rangle = 5 \cdot 0.01 + 0 \cdot (-0.02) + 0 \cdot 0.03 = 0.05 $$

##### Verification via finite difference 
  
$$ f(x) = 5 \cdot 1 = 5 $$

$$ x_{new} = x + dx = \begin{bmatrix} 1.01 & 1.98 & 3.03 \end{bmatrix} $$

$$ f(x_{new}) = 5 \cdot 1.01 = 5.05 $$

$$ f(x_{new}) - f(x) = 0.05 = df $$

$$ \diamond \diamond \diamond $$

### Example 2

##### Definition of the function

$$ f:\mathbb{R}^{n} \to \mathbb{R} $$

$$ f(x) = {||x||_2}^2 = \sum_{i} x_{i}^2 $$
##### Partial derivatives

$$ \frac{\partial f}{\partial x_i} = 2x_i $$
##### Gradient

$$ \nabla f(x) = \begin{bmatrix} \frac{\partial f}{\partial x_1} & \frac{\partial f}{\partial x_2} & \cdots & \frac{\partial f}{\partial x_n} \end{bmatrix} = \begin{bmatrix} 2x_1 & 2x_2 & \cdots & 2x_n \end{bmatrix} = 2x $$
##### Differential

$$ df = \langle \nabla f(x), dx \rangle = 2 \sum_{i} x_i dx_i $$
##### Numerical example

$$ x = \begin{bmatrix} 1 & 2 & 3\end{bmatrix} $$

$$ dx = \begin{bmatrix} 0.01 & -0.02 & 0.03 \end{bmatrix} $$

$$ \nabla f(x) = \begin{bmatrix} 2 & 4 & 6 \end{bmatrix} $$

$$ df = \langle \nabla f(x), dx \rangle = 2 \cdot 0.01 + 4 \cdot (-0.02) + 6 \cdot 0.03 = 0.12 $$
##### Verification via finite difference

$$ f(x) = 1^2 + 2^2 + 3^2 = 14 $$

$$ x_{new} = x + dx = \begin{bmatrix} 1.01 & 1.98 & 3.03 \end{bmatrix} $$

$$ f(x_{new}) = 1.01^2 + 1.98^2 + 3.03^2 = 14.12 $$

$$ f(x_{new}) - f(x) = 0.12 = df $$

$$ \diamond \diamond \diamond $$

### Example 3

##### Definition of the function

$$ f:\mathbb{R}^{n \times n} \to \mathbb{R} $$

$$ f(x) = tr(x) = \sum_{i} x_{ii} $$
##### Partial derivatives

$$ \frac{\partial f}{\partial x_{ii}} = \begin{dcases} 1 & \text{if } i = j \\ 0 & \text{otherwise} \end{dcases} $$

##### Gradient

$$ \nabla f(x) = \begin{bmatrix} \frac{\partial f}{\partial x_{11}} & \frac{\partial f}{\partial x_{12}} & \cdots & \frac{\partial f}{\partial x_{1n}} \\ \frac{\partial f}{\partial x_{21}} & \frac{\partial f}{\partial x_{22}} & \cdots & \frac{\partial f}{\partial x_{2n}} \\ \vdots & \vdots & \ddots & \vdots \\ \frac{\partial f}{\partial x_{n1}} & \frac{\partial f}{\partial x_{n2}} & \cdots & \frac{\partial f}{\partial x_{nn}} \end{bmatrix} = \begin{bmatrix} 1 & 0 & \cdots & 0 \\ 0 & 1 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & 1 \end{bmatrix} = I_n $$

##### Differential

$$ df = \langle \nabla f(x), dx \rangle = \langle I_n, dx \rangle = \sum_{i} dx_{ii} $$

##### Numerical example

$$ x = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} $$

$$ dx = \begin{bmatrix} 0.01 & -0.02 \\ 0.03 & 0.04 \end{bmatrix} $$

$$ \nabla f(x) = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} $$

$$ df = \langle \nabla f(x), dx \rangle = 0.01 + 0.04 = 0.05 $$
##### Verification via finite difference

$$ f(x) = 1 + 4 = 5 $$

$$ x_{new} = x + dx = \begin{bmatrix} 1.01 & 1.98 \\ 3.03 & 4.04 \end{bmatrix} $$

$$ f(x_{new}) = 1.01 + 4.04 = 5.05 $$

$$ f(x_{new}) - f(x) = 0.05 = df $$

$$ \diamond \diamond \diamond $$

### Example 4

##### Definition of the function

$$ f:\mathbb{R}^{n \times m} \times \mathbb{R}^{m \times n} \to \mathbb{R}^{n \times n} $$

$$ f(A, B) = tr(AB) = \sum_{i} (AB)_{ii} $$

The trace is a scalar, so this is a valid scalar‑valued function even though it depends on two matrices of different shapes.

##### Partial derivatives

$$ \frac{\partial f}{\partial A_{ij}} = B_{ji} $$

$$ \frac{\partial f}{\partial B_{ij}} = A_{ji} $$

##### Gradient

$$ \nabla_Af = B^T \in \mathbb{R}^{n \times m} $$

$$ \nabla_Bf = A^T \in \mathbb{R}^{m \times n} $$

$$ \nabla f(A, B) = \left( \begin{bmatrix} \frac{\partial f}{\partial A_{11}} & \cdots & \frac{\partial f}{\partial A_{1m}} \\ \vdots & \ddots & \vdots \\ \frac{\partial f}{\partial A_{n1}} & \cdots & \frac{\partial f}{\partial A_{nm}} \end{bmatrix}, \begin{bmatrix} \frac{\partial f}{\partial B_{11}} & \cdots & \frac{\partial f}{\partial B_{1n}} \\ \vdots & \ddots & \vdots \\ \frac{\partial f}{\partial B_{m1}} & \cdots & \frac{\partial f}{\partial B_{mn}} \end{bmatrix} \right) = (B^T, A^T) $$

The gradient always has the same shape as the argument. This is required by the Frobenius inner product in the differential.

##### Differential

$$ d(AB) = (dA)B + A(dB) $$

$$ df = tr(d(AB)) = tr((dA)B) + tr(A(dB)) $$

$$ tr(XY) = \langle X^T, Y \rangle $$

$$ df = \langle B^T, dA \rangle + \langle A^T, dB \rangle = \langle \nabla_Af, dA \rangle + \langle \nabla_Bf, dB \rangle $$

##### Numerical example

$$ A = \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 1 \end{bmatrix} $$

$$ B = \begin{bmatrix} 1 & 0 \\ 2 & 1 \\ 0 & 1 \end{bmatrix} $$

$$ AB = \begin{bmatrix} 5 & 2 \\ 2 & 2 \end{bmatrix} $$

$$ f(A, B) = 5 + 2 = 7 $$

$$ \nabla_Af = B^T = \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 1 \end{bmatrix} $$

$$ \nabla_Bf = A^T = \begin{bmatrix} 1 & 0 \\ 2 & 1 \\ 0 & 1 \end{bmatrix} $$

$$ dA = \begin{bmatrix} 0.01 & -0.02 & 0.0 \\ 0.0 & 0.03 & -0.01 \end{bmatrix} $$

$$ dB = \begin{bmatrix} 0.02 & 0.0 \\ -0.01 & 0.01 \\ 0.0 & 0.02 \end{bmatrix} $$

$$ df = \langle B^T, dA \rangle + \langle A^T, dB \rangle = -0.01 + 0.03 = 0.02 $$

##### Verification via finite difference

$$ A_{new} = A + dA = \begin{bmatrix} 1.01 & 1.98 & 0.0 \\ 0.0 & 1.03 & 0.99 \end{bmatrix} $$

$$ B_{new} = B + dB = \begin{bmatrix} 1.02 & 0.0 \\ 1.99 & 1.01 \\ 0.0 & 1.02 \end{bmatrix} $$

$$ f(A_{new}, B_{new}) = tr(A_{new}B_{new}) = 7.02 $$

$$ f(A_{new}, B_{new}) - f(A, B) = 0.02 \approx df $$

$$ \diamond \diamond \diamond $$
