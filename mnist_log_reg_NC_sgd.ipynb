```python
import torch as tr
from torchvision import datasets, transforms

import import_ipynb
from common import assert_eq, check_eq # type: ignore
from log_reg_NC_sgd_gradient import logistic_regression_nc_sgd_gradient as logistic_regression_nc # type: ignore


def test_mnist(s: int, t: int, epochs: int, lr=0.1) -> float:
  """
    Train a logistic regression model on the MNIST dataset and test its accuracy.

    Args:
      s: number of samples to use for training
      t: number of samples to use for testing
      epochs: number of training epochs
      lr: learning rate

    Returns:
      Accuracy of the model on the test set.
  """

  to_tensor = transforms.ToTensor()

  train_dataset = datasets.MNIST(root='data', train=True, download=True, transform=to_tensor)
  test_dataset = datasets.MNIST(root='data', train=False, download=True, transform=to_tensor)

  samples_train = train_dataset.data[:s].float() / 255.0
  assert_eq(samples_train.shape, (s, 28, 28))

  samples_train = tr.reshape(samples_train, (samples_train.shape[0], -1))
  assert_eq(samples_train.shape, (s, 784))

  labels_train = train_dataset.targets[:s]
  assert_eq(labels_train.shape, (s,))

  labels_train = tr.reshape(labels_train, (-1, 1))
  assert_eq(labels_train.shape, (s, 1))

  (_, model_W, model_b) = logistic_regression_nc(samples_train, labels_train, epochs=epochs, lr=lr)

  samples_test = test_dataset.data[:t].float() / 255.0
  assert_eq(samples_test.shape, (t, 28, 28))

  samples_test = tr.reshape(samples_test, (samples_test.shape[0], -1))
  assert_eq(samples_test.shape, (t, 784))

  labels_test = test_dataset.targets[:t]
  assert_eq(labels_test.shape, (t,))

  labels_test = tr.reshape(labels_test, (-1, 1))
  assert_eq(labels_test.shape, (t, 1))

  total = 0
  correct = 0

  for (sample, label) in zip(samples_test, labels_test):
      output = sample @ model_W + model_b

      actual_label = tr.argmax(output).item()
      expected_label = label.item()

      total += 1
      correct += check_eq(actual_label, expected_label)

  return correct / total


if __name__ == "__main__":
    #
    # May take a while to run, but should be above 90% accuracy.
    #

    print(test_mnist( 5000, 100,  200, 0.1))  # ~0.86
    print(test_mnist(10000, 200,  400, 0.1))  # ~0.86
    print(test_mnist(20000, 300,  600, 0.1))  # ~0.92
    print(test_mnist(30000, 400,  800, 0.1))  # ~0.92
    print(test_mnist(40000, 500, 1000, 0.1))  # ~0.92
    print(test_mnist(50000, 600, 1000, 0.1))  # ~0.92  
```
