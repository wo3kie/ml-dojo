### Vector inner product  

$$ \langle u, v \rangle = \sum_i u_i v_i = u^T \cdot v \quad u, v \in \mathbb{R}^n $$

```python
u = tr.tensor([1, 2, 3])
v = tr.tensor([4, 5, 6])
assert_eq(tr.dot(u, v), 1*4 + 2*5 + 3*6)
```

### Matrix inner product, aka Frobenius inner product

$$ \langle U, V \rangle_{F} = \sum_{i,j} U_{ij} V_{ij} = \text{tr}(U^T V) \quad U, V \in \mathbb{R}^{m \times n}$$

```python
U = tr.tensor([[1, 2], [3, 4]])
V = tr.tensor([[5, 6], [7, 8]])
assert_eq(tr.sum(U * V), 1*5 + 2*6 + 3*7 + 4*8)

UF = U.flatten()
VF = V.flatten()
assert_eq(tr.dot(UF, VF), 1*5 + 2*6 + 3*7 + 4*8)
```

### Matrix inner product properties

$$ (1) \quad \Big\langle U, AV \Big\rangle_{F} = 
\text{tr}\Big(U^T \, AV\Big) =
\text{tr}\Big((A^TU)^T \, V\Big) = 
\Big\langle A^TU, V \Big\rangle_{F} $$

$$ \\[2em] $$

$$ (2) \quad \Big\langle U, V\!A \Big\rangle_{F} = 
\text{tr} \Big( U^T \,\, V\!A \Big) =
\text{tr} \Big( AU^T \,\, V \Big) = 
\Big\langle U\!A^T, V \Big\rangle_{F} $$

### Numerical example

$$ z=xW $$
$$ p=\frac{e^z}{1+e^z} $$
$$ L=-y\log p - (1-y)\log(1-p) $$

$$ \diamond \diamond \diamond $$

$$ \frac{\partial z}{\partial x} = W $$

$$ \frac{\partial z}{\partial W} = x $$

$$ \frac{\partial L}{\partial z} = (p-y) $$

$$ 
\frac{\partial L}{\partial x} = 
\frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial x} \stackrel{(2)}{=}  
(p-y) \cdot W^T
$$

$$ 
\frac{\partial L}{\partial W} = 
\frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial W} \stackrel{(1)}{=}  
x^T \cdot(p-y) 
$$

```python
S = 5   # Samples
FI = 3  # Features in
FO = 2  # Features out

x = tr.randn(size=(S, FI), requires_grad=True, dtype=tr.float32)
W = tr.randn(size=(FI, FO), requires_grad=True, dtype=tr.float32)
y = tr.randint(low=0, high=2, size=(S, FO), dtype=tr.float32)

z = x @ W
p = tr.exp(z) / (1 + tr.exp(z))
L = -y * tr.log(p) - (1-y) * tr.log(1-p)

dL_dz = (p - y)
assert_eq(dL_dz.shape, z.shape)

# RuntimeError: mat1 and mat2 shapes cannot be multiplied (5x2 and 3x2)
# dL_dx = dL_dz @ W
dL_dx = dL_dz @ W.t()
assert_eq(dL_dx.shape, x.shape)

# mat1 and mat2 shapes cannot be multiplied (5x2 and 5x3)
# dL_dW = dL_dz @ x
dL_dW = x.t() @ dL_dz
assert_eq(dL_dW.shape, W.shape)
```
